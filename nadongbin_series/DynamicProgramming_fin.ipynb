# 다이나믹 프로그래밍
* 다이나믹 프로그래밍은 메모리를 적절히 사용하여 수행 시간 효율성을 비약적으로 향상시키는 방법

* 이미 계산된 결과는 별도의 메모리 영역에 저장하여 다시 계산하지 않도록 한다.

* 다이나믹 프로그래밍의 구현은 일반적으로 탑다운과 보텀업으로 구성된다.

* 다이나믹 프로그래밍은 동적 계획법이라고도 부른다.

* 일반적인 프로그래밍 분야에서 동적이란?

    * 자료구조에서 동적 할당은 프로그램이 실행되는 도중에 실행에 필요한 메모리를 할당하는 기법

    * 반면에 다이나믹 프로그래밍에서 다이나믹은 별다른 의미 없이 사용된 단어임



# 다이나믹 프로그래밍의 조건
1. 최적 부분 구조

    * 큰 문제를 작은 문제로 나눌 수 있으며 작은 문제의 답을 모아서 큰 문제를 해결할 수 있다.

2. 중복되는 문제

    * 동일한 작은 문제를 반복적으로 해결해야 함

```python
# 다이나믹 프로그래밍이 사용되는 대표적인 경우
# 피보나치 수열
def fibo(x):
    if x == 1 or x == 2:
        return 1
    return fibo(x - 1) + fibo(x - 2)
fibo(10)
```

# 메모이제이션(Memoization)
* 메모이제이션은 다이나믹 프로그래밍을 구현하는 방법 중 하나
* 한 번 계산한 결과를 메모리 공간에 메모하는 기법

    * 같은 문제를 다시 호출하면 메모했던 결과를 그대로 가져옴
    * 값을 기록해 놓는다는 점에서 캐싱이라고도 함

# 탑다운 VS 바텀업
* 탑다운(메모이제이션)방식은 하향식이라고도 하며 보텀업 방식은 상향식이라고도 한다.
* 다이나믹 프로그래밍의 전형적인 형태는 보텀업 방식

    * 결과 저장용 리스트는 DP 테이블이라고 부른다
* 엄밀히 말하면 메모이제이션은 이전에 계산된 결과를 일시적으로 기록해 놓는 넓은 개념을 의미
    * 따라서 메모이제이션은 다이나믹 프로그래밍에 국한된 개념은 아니다.
    * 한 번 계산된 결과를 담아 놓기만 하고 다이나믹 프로그래밍을 위해 활요하지 않을 수도 있다.

```python
# 다이나믹 프로그래밍으로 구현
# 피보나치 수열은
# 큰 문제를 작은 문제로 나눌 수 있고
# 동일한 작은 문제를 반복적으로 해결할 수도 있음
x = int(input())
fibo = [0]*x
fibo[0] = 1
fibo[1] = 1
for i in range(2, x):
    fibo[i] = fibo[i-2] + fibo[i-1]
print(fibo[-1])
    
```

# 다이나믹 프로그래밍 VS 분할 정복
* 다이나믹 프로그래밍과 분할 정복은 모두 최적 부분 구조를 가질 때 사용할 수 있다
    * 큰 문제를 작은 문제로 나눌 수 있으며 작은 문제의 답을 모아서 큰 문제를 해결할 수 있는 상황
* 다이나믹 프로그래밍과 분할 정복의 차이점은 부분 문제의 중복입니다.
    * 다이나믹 프로그래밍 문제에서는 각 부분 문제들이 서로 영향을 미치며 부분 문제가 중복된다.
    * 분할 정복 문제에서는 동일한 부분 문제가 반복적으로 계산되지 않는다.

# 다이나믹 프로그래밍 접근법
* 주어진 문제가 다이나믹 프로그래밍 유형임을 파악
* 가장 먼저 그리디, 구현, 완점 탐색 등의 아이디어로 문제를 해결할 수 있는지 검토
    * 다른 알고리즘으로 풀이 방법이 떠오르지 않으면 다이나믹 프로그래밍 고려
* 일단 재귀함수로 비효율적인 완점 탐색 프로그램을 작성한 뒤에 (탑다운) 작은 문제에서 구한 답이 큰 문제에서 그대로 사용될 수 있으면, 코드를 개선하는 방법을 사용
* 일반적으로 코딩 테스트 수준에서는 기본 유형의 다이나믹 프로그래밍 문제가 출제되는 경우가 많음

```python
# 문제 1 : 개미전사
n = 4 
k = [1, 3, 1, 5]
idx = 0
get_point = k[0]
rest_point = 0

while ((idx+1) != n-1) and ((idx+1) != n):
    if get_point + k[idx+2] >= rest_point + k[idx+1]:
        get_point, rest_point = get_point + k[idx+2], k[idx+1]
        idx = idx + 2
    else:
        get_point, rest_point = rest_point + k[idx+1], get_point
        idx = idx + 1

print(get_point)
```

```python
 # 문제 1로 만들기
 # 5로 나누어 떨어지면 5로 나눔
 # 3으로 나누어 떨어지면 3으로 나눔
 # 2로 나누어 떨어지면 2로 나눔
 # 1을 뺀다


# 그리디로 풀고자 했는데 잘못 품 >> 틀림
count = 0
def calculator(x):
    global count
    if x == 1: 
        print(count)
        return x
    
    if x % 5 == 0:
        count +=1
        x = calculator(x / 5)
    elif x % 5 == 1:     
        count +=2
        x = calculator((x-1) / 5)
    elif x % 5 == 2:
        count += 1
        x = calculator(x / 2)
    elif x % 5 == 3:
        count +=1
        x = calculator(x / 3)
    elif x % 5 == 4:
        count +=2
        x = calculator((x-1) / 3)
    
calculator(10000)
```

```python
# 문제 : 효율적인 화페 구성
m = 7
# m = 15
n = [2, 3, 5]
# n = [2, 3]
d = [10010] * (m+1)
for i in n:
    d[i] = 1

for i in range(n[-1], m+1):
    d[i] = min([d[i-x]+1 for x in n])

print(d[:m+1])
if d[m] >= 10001:  print(-1)
else:  print(d[m])
```

```python
# 문제 : 금광
for tc in range(int(input())):

    n, m = map(int, input().split())
    array = list(map(int, input().split()))

    dp = []
    index = 0
    for i in range(n):
        dp.append(array[index:index + m])
        index += m
    for j in range(1, m):
        for i in range(n):
            if i == 0: left_up = 0
            else: left_up = dp[i - 1][j - 1]

            if i == n- 1: left_down = 0
            else: left_down = dp[i + 1][j - 1]

            left = dp[i][j-1]

            dp[i][j] = dp[i][j] + max(left_up, left_down, left)
    result = 0
    for i in range(n):
        result = max(result, dp[i][m-1])
    print(result)

```

```python
# 문제 병사 배치하기
n = 7
power = [15, 11, 4, 8, 5, 2, 4]
dp = []
for i in range(n):
    temp = [power[i]]
    for j in range(i, n):
        if temp[-1] > power[j]:
            temp.append(power[j])
        else:
            if len(temp) == 1:
                pass

            elif temp[-2] > power[j]:
                temp[-1] = power[j]
    print(temp)
    dp.append(len(temp))
print(n - max(dp))
```
